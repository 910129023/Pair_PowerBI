```python
# Libraries 
# -----------------------------------------------------------------------
import pandas as pd
import numpy as np
import warnings
import sys
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
# pd.set_option('display.max_rows', None)
warnings.filterwarnings("ignore")
```

```python
# Para cada csv:

    # 1º Importar el csv como df

    # 2º Crear columna Type

    # 3º Completar columna con valor "Red" "Rose" "Sparkling" "White" según corresponda

# 4º Unir dfs uno debajo del otro por filas, ya que comparten columnas

# 5º Una vez unidos, imputar en columna Year, el valor "N.V" - Definir método según % de representación.abs

# 6º Pasar tipo de dato Year a Int

# 7º Guardar a csv limpio para usar en Power BI
```

```python
# 1º Importar el csv como df

data_red = pd.read_csv("data/Red.csv")
data_rose = pd.read_csv("data/Rose.csv")
data_sparkling = pd.read_csv("data/Sparkling.csv")
data_white = pd.read_csv("data/White.csv")
```

```python
# 2º Crear columna Type
# 3º Completar columna con valor "Red" "Rose" "Sparkling" "White" según corresponda

data_red.insert(1,"Type","Red")
data_rose.insert(1,"Type","Rose")
data_sparkling.insert(1,"Type","Sparkling")
data_white.insert(1,"Type","White")
```

```python
# 4º Unir dfs uno debajo del otro por filas, ya que comparten columnas

df_final = pd.concat([data_red,data_rose,data_sparkling,data_white], axis=0,join='outer',ignore_index=True)
df_final.sample(30)
```

```python
# 5º Una vez unidos, imputar en columna Year, el valor "N.V" - Definir método según % de representación.abs

# Reemplazar nulos con la moda

moda = df_final[["Year"]].mode()
df_final[["Year"]] = df_final[["Year"]].replace("N.V.", moda.iloc[0])

# Comprobamos que se reemplace correctamente
df_final["Year"].value_counts()
```

```python
# 6º Pasar tipo de dato Year a Int
df_final["Year"] = df_final["Year"].astype(int)


# Comprobamos que el tipo de dato se cambie correctamente 
df_final["Year"].dtype
```

```python
# 7º Guardar a csv limpio para usar en Power BI

df_final.to_csv("data/wines_data_final.csv")
```
