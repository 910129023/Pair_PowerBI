```python
data_sparkling.insert(1,"Type","Sparkling")
data_sparkling.head()
```

```python
data_white.insert(1,"Type","White")
data_white.head()
```

```python
df_final = pd.concat([data_red,data_rose,data_sparkling,data_white], axis=0,join='outer',ignore_index=True)
df_final.sample(30)
```

```python
df_final.shape
```

```python
# Ver % de "N.V" en columna Year (744/13834 = 5,3%)

df_final["Year"].value_counts()
```

```python
# No deja imputar porque no lo hace son strings (la columna es object), primero hay que convertir a 0 los "N.V" y luego ya los 0, convertirlos con replace porque la media con 0 se verá afectada.
```

```python
# Reemplazar nulos con la media 
media = df_final[["Year"]].mean()
df_final[["Year"]] = df_final[["Year"]].fillna(media)
```

```python
df_final["Year"] = df_final["Year"].astype(int)
```
